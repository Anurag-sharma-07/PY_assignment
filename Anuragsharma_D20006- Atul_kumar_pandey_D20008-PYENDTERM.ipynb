```python
# Q1 
def Histogram(data):                              # Defining functions which takes one argument
    num = data.select_dtypes(include=[np.number]) # creating a subset of data which will contain columns that only have numeric data type
    for col in num:
        ax=num[col].hist(bins=20,grid=False,edgecolor="black")    #generating histogram
        ax.set(xlabel=col, ylabel='Frequency')
        ax.set_title('Histogram of'+' '+col)
        ax.figure.savefig(col)                                   # saving graphs in png format
        plt.show()
        plt.close()                              # Close the current figure window
```

```python
Histogram(cars)
```

```python
Histogram(emp)
```

```python
def Histogram_sns(data):
    column = data.columns.tolist()
    num = data.select_dtypes(include=[np.number])
    for col in num:
        sns.displot(num[col])                          # histogram using displot from searborn
        plt.savefig(col)
        plt.show()
        plt.close()
```

```python
# Q2
def graph(data):
    num = data.select_dtypes(include=[np.number])   # creating a subset of data which will contain columns that only have numeric data type
    obj = data.select_dtypes(exclude=[np.number])   # creating a subset of data which will contain columns that do not have numeric data type
    for col in num:                                 #  A for loop will run through arbitrary variable col
        ax=num[col].hist(bins=20,grid=False,edgecolor="black")
        ax.set(xlabel=col, ylabel='Frequency')       # histogram will be generated for columns/ Variable that has numeric value   
        ax.set_title('Histogram of'+' '+col)
        ax.figure.savefig(col)
        plt.show()
        plt.close()
        ay=sns.boxplot(x=col,data=num)              # Box-plot will be generated for columns/ Variable that has numeric value using sea born library
        ay.set_title('Boxplot of'+' '+col)
        ay.figure.savefig(col)
        plt.show()
        plt.close()
    for column in obj:
        az=sns.countplot(x=column,data=obj)       # Bar/count-plot will be generated for columns/ Variable that has categorical value using sea born library
        az.set(xlabel=column, ylabel='Frequency')
        az.set_title('Barplot of'+' '+column)
        az.figure.savefig(column)                 # saving graphs in png format
        plt.show()
        plt.close()                             # Close the current figure window
```

```python
graph(cars)
```

```python
graph(emp)
```
